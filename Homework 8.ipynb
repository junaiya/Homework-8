Question 1

a) A Classification Decision Tree addresses problems where the goal is to categorize data into predefined classes based on input features. It is particularly useful for problems requiring clear decision-making and interpretability. For example, it can classify emails as "spam" or "not spam" based on text content, identify loan applicants as "approved" or "denied" using financial data, or predict whether a tumor is "benign" or "malignant" based on medical attributes. These trees are highly applicable in domains like finance, healthcare, marketing, and cybersecurity, where understanding the reasoning behind predictions is as important as the predictions themselves.

b) A **Classification Decision Tree** makes predictions by dividing data into branches based on feature values, ultimately assigning each data point to a specific category or class at the leaves. It uses thresholds and splits to group similar data points, making discrete predictions like "yes" or "no." 

In contrast, **Multiple Linear Regression** predicts a continuous outcome by fitting a linear equation to the data, combining feature values and their coefficients. For example, it estimates numerical values like house prices or sales revenue based on predictors. While decision trees classify into distinct groups, regression focuses on predicting a continuous numeric outcome.

Question 2

1. Accuracy is most appropriately used in real-world scenarios where the cost of false positives and false negatives is relatively low and the dataset is balanced, meaning the classes occur in roughly equal proportions. For example, it is suitable for quality control in manufacturing, where the goal is to identify defective products, assuming defects are neither too rare nor too frequent. Similarly, accuracy works well in customer feedback sentiment analysis, where the distribution of positive, neutral, and negative sentiments is reasonably even. However, it becomes less reliable in cases with class imbalances, such as medical diagnoses for rare diseases, where other metrics like precision or recall might be more appropriate.


2. Sensitivity, or recall, is most appropriately used in scenarios where it is crucial to identify as many actual positive cases as possible, even at the cost of some false positives. This is particularly important in applications like medical diagnostics for serious conditions (e.g., cancer screening), where missing a true positive could have severe consequences. It is also vital in fraud detection systems, where the goal is to flag all potentially fraudulent transactions, even if some legitimate transactions are incorrectly flagged, as these can be further reviewed. Sensitivity ensures that critical cases are not overlooked.

3. Specificity is most appropriately used in scenarios where correctly identifying true negatives is critical, and minimizing false positives is a priority. For instance, in medical diagnostics for conditions with serious psychological or financial implications from a false positive (e.g., HIV testing), high specificity ensures that individuals without the condition are not mistakenly identified as positive. Similarly, in spam email detection, high specificity reduces the likelihood of legitimate emails being marked as spam, ensuring critical communications are not lost. Specificity is crucial when the cost of false positives outweighs that of false negatives.

4. Precision is most appropriately used in scenarios where the focus is on the accuracy of positive predictions, and false positives have a significant cost or impact. For example, in email spam filters, high precision ensures that flagged emails are genuinely spam, minimizing the risk of misclassifying important emails as spam. Similarly, in criminal justice systems using predictive models, high precision ensures that individuals flagged for risk are indeed likely to exhibit the predicted behavior, reducing unnecessary legal or social consequences for wrongly flagged individuals. Precision is critical in applications where false positives must be minimized.

```python
import pandas as pd

file_path = "amazonbooks.csv"
ab = pd.read_csv(file_path, encoding="ISO-8859-1")

columns_to_drop = ['Weight_oz', 'Width', 'Height']
ab_reduced = ab.drop(columns=columns_to_drop)

ab_reduced_noNaN = ab_reduced.dropna().copy()

ab_reduced_noNaN['Pub year'] = ab_reduced_noNaN['Pub year'].astype(int)
ab_reduced_noNaN['NumPages'] = ab_reduced_noNaN['NumPages'].astype(int)
ab_reduced_noNaN['Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')

ab_reduced_noNaN.info(), ab_reduced_noNaN.head()
```

Question 4

```python
from sklearn.model_selection import train_test_split

random_seed = 42

ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(
    ab_reduced_noNaN, test_size=0.2, random_state=random_seed
)

print(f"Training set size: {len(ab_reduced_noNaN_train)} observations")
print(f"Testing set size: {len(ab_reduced_noNaN_test)} observations")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

y = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']  
X = ab_reduced_noNaN_train[['List Price']] 

clf = DecisionTreeClassifier(max_depth=2, random_state=random_seed)
clf.fit(X, y)

tree.plot_tree(clf, feature_names=['List Price'], class_names=['Paper', 'Hard'], filled=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
tree.plot_tree(clf, feature_names=['List Price'], class_names=['Paper', 'Hard'], filled=True)
plt.show()
```

Question 5

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import graphviz
from sklearn import tree
from IPython.display import FileLink

X = ab_reduced_noNaN_train[['NumPages', 'Thick', 'List Price']]
y = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']

clf2 = DecisionTreeClassifier(max_depth=4, random_state=42)
clf2.fit(X, y)

plt.figure(figsize=(12, 10))
tree.plot_tree(
    clf2,
    feature_names=['NumPages', 'Thick', 'List Price'],
    class_names=['Paper', 'Hard'],
    filled=True,
    rounded=True
)
plt.title("Decision Tree Visualization (Matplotlib)")
plt.show()

dot_data2 = tree.export_graphviz(
    clf2,
    out_file=None,
    feature_names=['NumPages', 'Thick', 'List Price'],
    class_names=['Paper', 'Hard'],
    filled=True,
    rounded=True,
    special_characters=True
)

graph2 = graphviz.Source(dot_data2)
graph2.render("decision_tree_clf2", format="pdf") 

print("Download the decision tree PDF:")
display(FileLink("decision_tree_clf2.pdf"))

X_test = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']

y_pred = clf2.predict(X_test)

print(classification_report(y_test, y_pred, target_names=['Paper', 'Hard']))
```

Question 6

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import numpy as np

def compute_metrics(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    TN, FP, FN, TP = cm.ravel()  
    
    sensitivity = TP / (TP + FN) if (TP + FN) > 0 else 0 
    specificity = TN / (TN + FP) if (TN + FP) > 0 else 0  
    accuracy = (TP + TN) / (TP + FP + FN + TN)
    
    return {
        "Sensitivity": np.round(sensitivity, 3),
        "Specificity": np.round(specificity, 3),
        "Accuracy": np.round(accuracy, 3)
    }, cm

y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']

y_pred_clf = clf.predict(ab_reduced_noNaN_test[['List Price']])
y_pred_clf2 = clf2.predict(ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']])

metrics_clf, cm_clf = compute_metrics(y_test, y_pred_clf)
print("Metrics for clf:")
print(metrics_clf)

ConfusionMatrixDisplay(confusion_matrix=cm_clf, display_labels=['Paper', 'Hard']).plot(cmap="Blues")
plt.title("Confusion Matrix for clf")
plt.show()

metrics_clf2, cm_clf2 = compute_metrics(y_test, y_pred_clf2)
print("Metrics for clf2:")
print(metrics_clf2)

ConfusionMatrixDisplay(confusion_matrix=cm_clf2, display_labels=['Paper', 'Hard']).plot(cmap="Blues")
plt.title("Confusion Matrix for clf2")
plt.show()
```
